```python
fig, ax = plt.subplots()
x = []
for i in range(len(cpp_result)):
    for j in range(5): #各5セット繰り返している
        x.append(i+1) #添え字が0始まりであることに注意
y = cpp_result

ax.set_xlabel('N')
ax.set_ylabel('time [ms]')
plt.title('Array multiple(C++)')

ax.scatter(x,y)
plt.show()
```
